```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
y = df['Outcome']
y
```

```python
x = df.drop('Outcome', axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 100)
x_train
```

```python
y_train
```

###**Bagging Regressor**

```python
from sklearn.ensemble import BaggingRegressor
```

```python
model = BaggingRegressor()
model.fit(x_train, y_train)
```

```python
y_br_train_pred = model.predict(x_train)
y_br_test_pred = model.predict(x_test)
```

```python
print(y_br_train_pred)
```

```python
print(y_br_test_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

br_train_mse = mean_squared_error(y_train, y_br_train_pred)
br_train_r2 = r2_score(y_train, y_br_train_pred)

br_test_mse = mean_squared_error(y_test, y_br_test_pred)
br_test_r2 =  r2_score(y_test, y_br_test_pred)
```

```python
br_results = pd.DataFrame(["Bagging regression",br_train_mse, br_train_r2, br_test_mse, br_test_r2]).transpose()
br_results.columns = ['Method', 'Traning Mse', 'Traning R2', 'Test Mse', 'Test R2']
```

```python
br_results
```
